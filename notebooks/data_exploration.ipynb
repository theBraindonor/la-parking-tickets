# Data Exploration
A basic statistical analysis of the LA Parking Data.  Preprocessing has been performed prior to this analysis.

```python
__author__ = "John Hoff"
__email__ = "john.hoff@braindonor.net"
__copyright__ = "Copyright 2019, John Hoff"
__license__ = "Creative Commons Attribution-ShareAlike 4.0 International License"
__version__ = "1.0.0"

import os
if os.getcwd().endswith('notebooks'):
    os.chdir('..')

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from data import *
```

```python
data_frame = pd.read_csv('data_scratch/preprocessed_citations.csv', dtype='float')
```

### Violation Class Distribution

```python
plt.figure(figsize=(10,5))
counts = data_frame['violation'].value_counts().sort_index()
counts.index = violation_codes.keys()
counts.plot('bar')
plt.xlabel('Violation')
plt.show()
counts
```

### Month Class Distribution

```python
plt.figure(figsize=(10,5))
counts = data_frame['month'].value_counts().sort_index()
counts.plot('bar')
plt.xlabel('Month')
plt.show()
counts
```

### Weekday Class Distribution

```python
plt.figure(figsize=(10,5))
counts = data_frame['weekday'].value_counts().sort_index()
counts.plot('bar')
plt.xlabel('Weekday')
plt.show()
counts
```

### Hour Class Distribution

```python
plt.figure(figsize=(10,5))
counts = data_frame['hour'].value_counts().sort_index()
counts.plot('bar')
plt.xlabel('Hour')
plt.show()
counts
```

### Vehicle State Class Distribution

```python
plt.figure(figsize=(15,5))
counts = data_frame['state'].value_counts().sort_index()
counts.index = state_codes.keys()
counts.plot('bar')
plt.xlabel('Vehicle State')
plt.show()
counts
```

### Vehicle Make Class Distribution

```python
plt.figure(figsize=(15,5))
counts = data_frame['make'].value_counts().sort_index()
counts.index = make_codes.keys()
counts.plot('bar')
plt.xlabel('Vehicle Make')
plt.show()
counts
```

### Vehicle Color Class Distribution

```python
plt.figure(figsize=(10,5))
counts = data_frame['color'].value_counts().sort_index()
counts.index = color_codes.keys()
counts.plot('bar')
plt.xlabel('Vehicle Color')
plt.show()
counts
```
